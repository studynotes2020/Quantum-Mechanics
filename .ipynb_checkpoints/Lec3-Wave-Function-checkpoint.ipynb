### 2. $P(x) = |\Psi(x)|^2$. Norm squared of a complex number is always a real non-negative number. The equation determines the probability density that the object in state $\Phi(x)$ will be found found at position x

Another way to state with normalization $$P(x) = \frac{|\Psi(x)|^2}{\int_{\forall x} |\Psi|^2}$$

$P(x, x + dx) = P(x)\ dx = |\Psi(x)|^2$. Probability density means probability of finding particles between x and dx
    
$\int_{\forall x} dx P(x) = 1 \implies \int_{\forall x} dx |\Psi(x)|^2 = 1$

Dimension of $\Psi(x)$ is $\frac{1}{\sqrt{length}}$

- Plot P(x) for delta function: Big spike at x1. At x1, the uncertainty $\Delta x$ is small

The function has no definite wavelength, and relation of wavelength and  momentum $P = \hbar \frac{2\pi}{\lambda}$, so uncertainly of momentum is large

- Plot P(x) for delta function: Big spike at x2. At x2, the uncertainty $\Delta x$ is small

Similar argument, the uncertainly of momentum is large

- Plot P(x) for sine curve, $\Psi = e^{ik_1x}$. 

$|\Psi|^2 = |e^{ik_1x}|^2 = e^{-ik_1x}.e^{ik_1x} = 1$

Probability density function is 1 and independent of x

No information about where is particle. Uncertainty is very large

However, we are sure about the wavelength. So we know its momentum, $P = \hbar k_1$. This is a periodic function, so the uncertainty of momentum is small

- Plot higher frequency sine curve, $\Psi = e^{ik_2x}$

Probability density function is again constant

Similarly, No information about where is particle. Uncertainty is very large

Similary, we are sure about the wavelength. So we know its momentum, $P = \hbar k_2$. This is a periodic function, so the uncertainty of momentum is small

- Plot P(x) for an arbitrary curve. Uncertainty not giant, but not sure where is x

### 3. This is the most important postulate. Given two possible configurations of a quantum system, corresponding to two wave functions $\Psi_1(x)$ and $\Psi_2(x)$, the system can also be in a superposition of these 

$$\Psi(x) = \alpha \Psi_1(x) + \beta \Psi_2(x)$$

$\alpha, \beta$ are complex subject to normalize

- Take superposition for two delta functions

particle at somwhere in between on average. uncertainty of position is now in order x1 - x2

- Superposition for two sinusoid functions.

$$\Psi(x) = e^{ik_1x} + e^{ik_2x}$$

$P(x) \text{ vs } x$ is no longer constant, but there's an interference term

$P(x) = |\alpha \Psi_1 + \beta \Psi_2|^2 = |\alpha|^2 |\Psi_1|^2 + |\beta| |\Psi_2|^2 + \alpha^* \Psi_1^* \beta \Psi_2 + \alpha \Psi_1 \beta^* \Psi_2^*$ 

$= P_1 + P_2 + \text{Interference}$

The complex conjugate terms multiply and add is real but might be negative. However the total RHS is always positive. The wave functions add but not the probabilities

We have some information about position, and huge uncertainty on momentum

Side note: $|e^{ia} + e^{ib}|^2 = |e^{ia}|^2 |(1 + e^{i(b-a)})|^2 = \cos(b-a)$

## Fourier Transform

Theorem: Any function $f(x)$ can be built by superposition enough plane waves $e^{ikx}$

$$f(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^\infty dk\ \tilde f(k) e^{ikx}$$

$\lambda = \frac{2\pi}{k}$

Inverse fourier transform

$$\tilde f(k) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^\infty dx\  f(x) e^{-ikx}$$

Physical Version: Any wave function $\Psi(x)$ can be expressed as a superposition in the form 

$$\Psi(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^\infty dk\ \tilde \Psi(k) e^{ikx}$$

of states with definite momentum $p = \hbar k$

Fourier transform $\tilde \Psi(k)$ for delta function at x1 is a sinusoid with frequency x1. $P(k) = |\Psi(k)|^2$. Definite value of P means small uncertainty

Fourier transform $\tilde \Psi(k)$ for sinusoid is a delta function at its frequency k1. Similarly here, definite value of P means small uncertainty

Fourier transform for delta function at x1 and x2 is $e^{ikx_1} + e^{ikx_2}$

Fourier transform for a superimposed wave of frequency k1 and k2 is delta function at k1 and k2


### In short, from the result of fourier transform, we see the uncertainty relation between position and momentum
